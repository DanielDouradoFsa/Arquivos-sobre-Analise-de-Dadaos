```python
#Importando bibliotecas e abrindo arquivos necessários
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt # 3 
from sklearn.cluster import KMeans # 4
abas= pd.ExcelFile("input_planilhaCanalHolder_20200211.xlsx")
MA=abas.parse('Múltiplos Ações')
FA=abas.parse('Fundamentos Ações')
```

```python
#Merge entre as duas primeiras abas do arquivo input_planilhaCanalHolder_20200211.xlsx
uniaoAbas = pd.merge(MA, FA, on="Nome", how='outer', suffixes=('', '_y'))
uniaoAbas = uniaoAbas.drop(list(uniaoAbas.filter(regex='_y$')), axis=1)
```

```python
#Removendo espaços nos nomes das colunas, removendo uma coluna inutil e substituindo atributos vazios para NaN e os removendo
uniaoAbas.columns = [str(c).replace(' ', '_') for c in uniaoAbas.columns]
uniaoAbas.drop("Unnamed:_0",axis=1,inplace=True)
uniaoAbas.replace("-",np.nan,inplace=True)
dadosCompletos=uniaoAbas.dropna()
```

```python
numerics = ['int16', 'int32', 'int64', 'float16', 'float32', 'float64']
newdf = dadosCompletos.select_dtypes(include=numerics)
```

```python
#Com base em todos os tipos de Setores_Bovespa presentes no dataFrame foi calculado o numero Clusters
numClusters=0
for k in dadosCompletos["Setor_Bovespa"].unique():
    numClusters+=1
```

```python
#Clusterização sendo feita nos atributos numéricos da tabela
kmeans = KMeans(n_clusters=numClusters, random_state=0).fit(newdf)
newdf_clustered = kmeans.fit_predict(newdf)
```

```python
#Criando uma tabela resultado para avaliar se a clusterização conseguiu agrupar corretamente as instâncias similares
results=dadosCompletos[["Setor_Bovespa"]].copy()
results["clusterNumber"]=newdf_clustered
results.head(60)
```
